```python
import numpy as np
% matplotlib inline
import matplotlib.pyplot as plt
```

### 3D Vectors
Plotting a 3D vector

```python
a = np.array([1, 2, 8])
b = np.array([5, 6, 3])
```

```python
x, y, z = zip(a, b)
```

```python
from mpl_toolkits.mplot3d import Axes3D

s3Dplot = plt.subplot(111, projection='3d')
s3Dplot.scatter(x, y, z)
s3Dplot.set_zlim3d([0, 9])
s3Dplot.set_ylim3d([0, 9])
s3Dplot.set_xlim3d([0, 9])
plt.show()
```

```python
# A function to plot given list of 3D vectors with attached vertical line to define 3d space
def plot_vectors3D(ax, vectors3D, z0, **options):
    for v in vectors3D:
        x, y, z = v
        ax.plot([x,x], [y,y], [z0, z], color='gray', linestyle='dotted', marker='.')
    x_coord, y_coord, z_coord = zip(*vectors3D)
    ax.scatter(x_coord, y_coord, z_coord, **options)
    
```

```python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
subplot3d = plt.subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vectors3D(subplot3d, [a,b], 0, color=('r', 'b'))
```

#### Scaler Multiplication
When done, all elements are multiplied by the number
When viewed graphically, scaler multiplied vector results in the change of scale of the figure.

```python
t1 = np.array([2, 0.25])
t2 = np.array([2.5, 3.5])
t3 = np.array([1, 2])
x, y = zip(t1, t2, t3, t1)
# scaling by a factor k
k = 2.5
t1c = k * t1
t2c = k * t2
t3c = k * t3
plt.plot(x, y, 'c--', x, y, 'co')
plot_vector2d(t1, color="r")
plot_vector2d(t2, color="r")
plot_vector2d(t3, color="r")

# plot the scaled vector
x_c, y_c = zip(t1c, t2c, t3c, t1c)
plt.plot(x_c, y_c, 'b-', x_c, y_c, 'bo')

# plot the 2d vector

plot_vector2d(k * t1, color="b", linestyle=":")
plot_vector2d(k * t2, color="b", linestyle=":")
plot_vector2d(k * t3, color="b", linestyle=":")
plt.axis([0,9,0,9])
plt.grid()
plt.show()
t1, t2, t3, t1, x,y
```

#### Normalised Vector
A vector division by a scaler can be represented as a multiplication by the inverse of that scaler
$\dfrac{\textbf{u}}{\lambda} = \dfrac{1}{\lambda} \times \textbf{u}$

Normalised vector (non-null) ${\textbf{u}}$. Is the unit (1) vector that points in the same direction as ${\textbf{u}}$
Whish is equal to: $\hat{\textbf{u}} = \dfrac{\textbf{u}}{\left \Vert \textbf{u} \right \|}$

```python
import numpy.linalg as LA
v = np.array([3, 1])
plt.gca().add_artist(plt.Circle((0,0),1,color='c'))
plt.plot(0, 0, "ko")
plot_vector2d(v / LA.norm(v), color="k")
plot_vector2d(v, color="b", linestyle=":")
plt.text(0.3, 0.3, "$\hat{u}$", color="k", fontsize=18)
plt.text(1.5, 0.7, "$u$", color="b", fontsize=18)
plt.axis([-1.5, 5.5, -1.5, 3.5])
plt.grid()
plt.show()
```
